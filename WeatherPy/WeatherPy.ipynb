```python
#Importing all dependencies here
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import openweathermapy as ow
from pprint import pprint
#from api_keys import api_key
#Importing Python library to pull cities based on Latitude and Longitude
from citipy import citipy
```

```python
lat_lng = []
cities = []

lats = np.random.uniform(low=-45.000, high=45.000, size=1500)
lngs = np.random.uniform(low=-90.000, high=90.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    print(lat_lng)
    cities.append(citipy.nearest_city(lat_lng[0],lat_lng[1]))
    
print(len(lat_lngs.tolist()))
print(len(cities))
```
